```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a reference the CSV file desired
csv_path1 = "raw_data/city_data.csv"
csv_path2 = "raw_data/ride_data.csv"

# Read the CSV into a Pandas DataFrame
city_df = pd.read_csv(csv_path1)
city_df.head()

ride_df = pd.read_csv(csv_path2)
ride_df.head()

#create data frame to find avg fare and total rides per city
city_ride = ride_df.groupby("city")
avg_fare = city_ride['fare'].mean()
total_ride = city_ride['fare'].count()
ride_final = pd.DataFrame({'avg_fare':avg_fare,'total_rides':total_ride})

#create data frame to find driver count and city type for each city
city_driver= city_df.groupby("city")
total_drivers_df= city_driver['driver_count'].sum()
city_type= city_driver["type"].last()
city_summary_final = pd.DataFrame({'driver_count':total_drivers_df,'type':city_type})

#combine city and ride data frames into single table
final_city_summary = pd.merge(ride_final, city_summary_final, left_index=True,right_index=True, how="outer")
final_city_summary.head()
```

```python
#Pie Charts

#create data frame to find total fare and total rides per city
pie_city_ride = ride_df.groupby("city")
pie_total_fare = pie_city_ride['fare'].sum()
pie_total_ride = pie_city_ride['fare'].count()
pie_ride_final = pd.DataFrame({'total_fare':pie_total_fare,'total_rides':pie_total_ride})

#create data frame to find driver count and city type for each city
pie_city_driver= city_df.groupby("city")
pie_total_drivers_df= pie_city_driver['driver_count'].sum()
pie_city_type= pie_city_driver["type"].last()
pie_city_summary_final = pd.DataFrame({'driver_count':pie_total_drivers_df,'type':city_type})

#combine city and ride data frames into single table
pie_final_city_summary = pd.merge(pie_ride_final, pie_city_summary_final, left_index=True,right_index=True, how="outer")
pie_final_city_summary.set_index("type")
pie_final_city_summary.head()

```
