```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a reference the CSV file desired
csv_path1 = "raw_data/city_data.csv"
csv_path2 = "raw_data/ride_data.csv"

# Read the CSVs into a Pandas DataFrame
city_df = pd.read_csv(csv_path1)
city_df.head()

ride_df = pd.read_csv(csv_path2)
ride_df.head()

#create data frame to find avg fare and total rides per city
city_ride = ride_df.groupby("city")
avg_fare = city_ride['fare'].mean()
total_ride = city_ride['fare'].count()
ride_final = pd.DataFrame({'avg_fare':avg_fare,'total_rides':total_ride})

#create data frame to find driver count and city type for each city
city_driver= city_df.groupby("city")
total_drivers_df= city_driver['driver_count'].sum()
city_type= city_driver["type"].last()
city_summary_final = pd.DataFrame({'driver_count':total_drivers_df,'type':city_type})

#combine city and ride data frames into single table
final_city_summary = pd.merge(ride_final, city_summary_final, left_index=True,right_index=True, how="outer")
final_city_summary.head()
```

```python
# Create a bubble plot for each city type that graphs each of the three metrics were comparing.

urban_table = final_city_summary.loc[final_city_summary["type"]=="Urban"]
plt.scatter(urban_table['total_rides'],urban_table['avg_fare'],marker="o", facecolors="lightcoral", edgecolors='black', s=urban_table['driver_count']*5, alpha=0.75,label='Urban') 

suburban_table = final_city_summary.loc[final_city_summary["type"]=="Suburban"]
plt.scatter(suburban_table['total_rides'],suburban_table['avg_fare'],marker="o", facecolors="lightskyblue", edgecolors="black", s=suburban_table['driver_count']*5, alpha=0.75,label='Suburban') 

rural_table = final_city_summary.loc[final_city_summary["type"]=="Rural"]
plt.scatter(rural_table['total_rides'],rural_table['avg_fare'],marker="o", facecolors="gold", edgecolors="black",s=rural_table['driver_count']*5, alpha=0.75,label='Rural')
## Bubble Plot ## filtering the merged table to urban only 
 ##plotting each category, *S is multiplied by 5 to increase bubble size 




plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fares($)")
plt.show()
```
